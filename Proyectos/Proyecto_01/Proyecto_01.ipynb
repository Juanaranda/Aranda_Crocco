### Clase Profiler

````python
class Profiler():
    def __init__(self, data):
        self.data = data
        self.date = date.today().strftime("%d-%m-%Y")
        self.output_folder = f"EDA_{self.date}"


        # Crea la carpeta EDA con la fecha actual
        os.makedirs(self.output_folder, exist_ok=True)

# Summarize
    def summarize(self, columns_of_interest=None, outlier_threshold=3):
        """
        Genera un resumen de estad√≠sticas y caracter√≠sticas para las variables del conjunto de datos. Esto incluye el tipo de variable,
        el n√∫mero de valores √∫nicos, el porcentaje de valores nulos, el n√∫mero y porcentaje de valores cero (solo para variables num√©ricas),
        estad√≠sticas descriptivas y la detecci√≥n de outliers. 
        
        Guarda un archivo de resumen en la carpeta de salida.

        Par√°metros
        ----------
        columns_of_interest: list, default=None
            Lista de columnas a resumir. Si es None, se resumen todas las columnas.
        
        outlier_threshold: int, default=3
            Umbral para detectar outliers en variables num√©ricas (valor por encima o por debajo del umbral).
        """
        # Si no se especifican columnas de inter√©s, se resumen todas las columnas
        if columns_of_interest is None:
            columns_of_interest = self.data.columns

        summary_file = os.path.join(self.output_folder, "summary.txt")

        with open(summary_file, "w") as file:
            for column in columns_of_interest:
                file.write(f"Variable: {column}\n")

                # Tipo de variable
                variable_type = str(self.data[column].dtype)
                file.write(f"Tipo de Variable: {variable_type}\n")

                # N√∫mero de valores √∫nicos
                num_unique = self.data[column].nunique()
                file.write(f"N√∫mero de Valores √önicos: {num_unique}\n")

                # Porcentaje de valores √∫nicos
                pct_unique = (num_unique / len(self.data)) * 100
                file.write(f"Porcentaje de Valores √önicos: {pct_unique:.2f}%\n")

                # N√∫mero de valores nulos
                num_null = self.data[column].isnull().sum()
                file.write(f"N√∫mero de Valores Nulos: {num_null}\n")

                # Porcentaje de valores nulos
                pct_null = (num_null / len(self.data)) * 100
                file.write(f"Porcentaje de Valores Nulos: {pct_null:.2f}%\n")

                # Si la variable es num√©rica
                if pd.api.types.is_numeric_dtype(self.data[column]):
                    # N√∫mero de valores cero
                    num_zeros = (self.data[column] == 0).sum()
                    file.write(f"N√∫mero de Valores Cero: {num_zeros}\n")

                    # Porcentaje de valores cero
                    pct_zeros = (num_zeros / len(self.data)) * 100
                    file.write(f"Porcentaje de Valores Cero: {pct_zeros:.2f}%\n")

                    # Estad√≠sticas descriptivas
                    descriptive_stats = self.data[column].describe()
                    file.write("Estad√≠sticas Descriptivas:\n")
                    file.write(descriptive_stats.to_string() + "\n")

                    # Detectar outliers (valores que est√°n por encima o por debajo del umbral definido)
                    upper_threshold = descriptive_stats["75%"] + (outlier_threshold * (descriptive_stats["75%"] - descriptive_stats["25%"]))
                    lower_threshold = descriptive_stats["25%"] - (outlier_threshold * (descriptive_stats["75%"] - descriptive_stats["25%"]))
                    outliers = ((self.data[column] > upper_threshold) | (self.data[column] < lower_threshold)).sum()
                    file.write(f"N√∫mero de Outliers (umbral={outlier_threshold}): {outliers}\n")

                    # Si hay outliers, generar una alerta
                    if outliers > 0:
                        file.write("ALERTA: Se encontraron outliers en esta variable.\n")

                # Separador entre variables
                file.write("-" * 50 + "\n")

# Plot vars
    def plot_vars(self, columns_of_interest=None, top_n=10):
        """
        Genera visualizaciones para las variables del conjunto de datos. Puede incluir histogramas para variables num√©ricas
        y gr√°ficos de barras para variables categ√≥ricas. 
        
        Guarda visualizaciones en una carpeta especificada.

        Par√°metros
        ----------
        columns_of_interest: list
            Lista de columnas a visualizar. Si es None, se utilizar√°n todas las columnas.
        top_n: int
            N√∫mero de categor√≠as m√°s comunes a mostrar en gr√°ficos de barras para variables categ√≥ricas.
        """
        if columns_of_interest is None:
            columns_of_interest = self.data.columns
        plots_folder = os.path.join(self.output_folder, "plots")
        os.makedirs(plots_folder, exist_ok=True)

        for column in columns_of_interest:
            if column in self.data.select_dtypes(include=[np.number]).columns:
                self.plot_numeric_variable(column, plots_folder)
            elif column in self.data.select_dtypes(include=["object"]).columns:
                self.plot_categorical_variable(column, plots_folder, top_n)

    def plot_numeric_variable(self, column, plots_folder):
        """
        Genera visualizaciones para una variable num√©rica dada, incluyendo un histograma de densidad y una matriz de correlaci√≥n.

        Guarda las visualizaciones en la carpeta especificada.

        Par√°metros
        ----------
        column: str
            El nombre de la columna num√©rica a visualizar.
        
        plots_folder: str
            Ruta de la carpeta donde se guardar√°n las visualizaciones.
        """
        fig = px.histogram(self.data, x=column, title=f'Distribuci√≥n de densidad para {column}', nbins=20)
        fig.update_layout(xaxis_title=column, yaxis_title='Densidad')
        fig.update_xaxes(categoryorder="total descending")

        plot_file = os.path.join(plots_folder, f"{column}_density.pdf")
        pio.write_image(fig, plot_file, format="pdf")

        corr = self.data.select_dtypes(include=[np.number]).corr()
        fig = go.Figure(data=go.Heatmap(z=corr.values, x=corr.index, y=corr.columns, colorscale='viridis', zmin=-1, zmax=1))
        fig.update_layout(title=f'Correlaci√≥n entre {column} y las dem√°s variables num√©ricas')

        corr_plot_file = os.path.join(plots_folder, f"correlation_{column}.pdf")
        pio.write_image(fig, corr_plot_file, format="pdf")

    def plot_categorical_variable(self, column, plots_folder, top_n):
        """
        Genera visualizaciones para una variable categ√≥rica dada, incluyendo un histograma de las categor√≠as m√°s comunes y una matriz
        de coeficiente de Cramer para evaluar la asociaci√≥n con otras variables categ√≥ricas.

        Guarda visualizaciones en la carpeta especificada.

        Par√°metros
        ----------
        column: str
            El nombre de la columna categ√≥rica a visualizar.
        
        plots_folder: str
            Ruta de la carpeta donde se guardar√°n las visualizaciones.
        
        top_n: int
            N√∫mero de categor√≠as m√°s comunes a mostrar en el histograma.
        """
        top_categories = self.data[column].value_counts().head(top_n).reset_index()
        top_categories.columns = [column, 'Frecuencia']

        fig = px.bar(top_categories, x=column, y='Frecuencia', title=f'Histograma de las {top_n} categor√≠as m√°s comunes en {column}')
        fig.update_xaxes(categoryorder="total descending")
        fig.update_xaxes(title_text=column)
        fig.update_yaxes(title_text='Frecuencia')

        plot_file = os.path.join(plots_folder, f"{column}.pdf")
        pio.write_image(fig, plot_file, format="pdf")

        if column != self.data.select_dtypes(include=["object"]).columns[-1]:
            associations = pd.DataFrame(index=self.data.select_dtypes(include=["object"]).columns, columns=[column])
            for col1 in self.data.select_dtypes(include=["object"]).columns:
                if col1 != column:
                    confusion_matrix = pd.crosstab(self.data[col1], self.data[column])
                    chi2, _, _, _ = chi2_contingency(confusion_matrix)
                    n = self.data.shape[0]
                    phi2 = chi2 / n
                    r, k = confusion_matrix.shape
                    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
                    rcorr = r - ((r-1)**2)/(n-1)
                    kcorr = k - ((k-1)**2)/(n-1)
                    cramer_v = np.sqrt(phi2corr / (min((kcorr-1), (rcorr-1))))
                    associations.at[col1, column] = cramer_v

            fig = px.imshow(associations,
                x=associations.columns, y=associations.index,
                color_continuous_scale='viridis',
                labels=dict(x="Variable", y="Variable", color="Cramer V"),
                title=f'Coeficiente V de Cramer entre variables categ√≥ricas')

            cramer_plot_file = os.path.join(plots_folder, f"cramer_{column}.pdf")
            pio.write_image(fig, cramer_plot_file, format="pdf")

# Clean data
    def dtype_transform(self, data, max_categoric_value):
        """
        Transforma las columnas tipo number y object que tengan 10 o menos datos √∫nicos, en columnas categ√≥ricas.

        Par√°metros
        ----------
        data: Dataframe
            Proveniente de funci√≥n scale, este debe venir ya procesado por el m√©todo clean_data

        max_categoric_value: int, dafault=10 
            Cantidad m√°xima de valores √∫nicos que puede tener una columna para ser tratada como categ√≥rica.

        Retorna
        -------
        data: Dataframe
            Contiene las columnas transformadas.
        """
        num_cols = data.select_dtypes(include='number').columns

        # Obtener las columnas que no son num√©ricas
        non_numeric_cols = data.select_dtypes(exclude=['number']).columns

        # Iterar sobre las columnas no num√©ricas y tratar de convertirlas a num√©ricas
        for col in non_numeric_cols:
            try:
                data[col] = pd.to_numeric(data[col])
            except ValueError:
                continue

        obj_cols = data.select_dtypes(include='object').columns

        # Convertir columnas num√©ricas con 10 o menos valores √∫nicos a categ√≥ricas
        data[num_cols[data[num_cols].nunique() <= 10]] = data[num_cols[data[num_cols].nunique() <= 10]].astype('category')

        # Convertir columnas no num√©ricas con max_categoric_value o menos valores √∫nicos a categ√≥ricas
        data[obj_cols[data[obj_cols].nunique() <= max_categoric_value]] = data[obj_cols[data[obj_cols].nunique() <= max_categoric_value]].astype('category')

        return data

    def get_imputer(self, data):
        """
        Realiza la imputaci√≥n de valores nulos en un conjunto de datos.

        Los valores nulos en las columnas num√©ricas se imputan utilizando la estrategia de la media,
        mientras que en las columnas categ√≥ricas se utiliza la estrategia del valor m√°s frecuente.

        Par√°metros
        ----------
        data: DataFrame
            Conjunto de datos que se va a imputar.

        Retorna
        -------
        - data: DataFrame
            Contiene valores nulos imputados.
        """
        # Identificar las columnas num√©ricas y categ√≥ricas
        numeric_columns = data.select_dtypes(include='number').columns
        categorical_columns = data.select_dtypes(include='category').columns

        # Imputar columnas num√©ricas
        numeric_imputer = SimpleImputer(strategy='mean')
        data[numeric_columns] = numeric_imputer.fit_transform(data[numeric_columns])

        # Imputar columnas categ√≥ricas
        #categorical_imputer = SimpleImputer(strategy='most_frequent')
        #data[categorical_columns] = categorical_imputer.fit_transform(data[categorical_columns])
        for col in categorical_columns:
            if "No Value" not in data[col].cat.categories:
                data[col] = data[col].cat.add_categories("No Value")
            # Llenar valores nulos con "No Value"
            data[col].fillna('No Value', inplace=True)

        return data

    def clean_data(self, columns_of_interest=None, max_categoric_value=1000, null_handling_method="drop", column_name_to_split=None):
        """
        Realiza la limpieza y preprocesamiento de un conjunto de datos.

        Permite separar columnas no at√≥micas, eliminar duplicados y tratar los valores nulos seg√∫n la estrategia especificada.

        Los datos procesados se guardan en un archivo CSV en una carpeta llamda "clean_data" dentro del directorio de trabajo.

        Crea un transformer que puede ser entregado en m√©todos posteriorires dentro de un pipeline.

        Par√°metros
        ----------
        columns_of_interest: list
            Lista de columnas de inter√©s para el procesamiento. Si es None, se consideran todas las columnas.

        max_categoric_value: int, default=1000
            Cantidad m√°xima de valores √∫nicos que puede tener una columna para ser tratada como categ√≥rica.
        
        null_handling_method: str, {'drop', 'impute'}='drop'
            M√©todo para tratar los valores nulos.
                'drop': elimina filas con valores nulos.
                
                'impute': valores nulos n√∫mericos de las filas son imputados utilizando la estrategia de la media
                        y en los valores categ√≥ricos se utiliza la estrategia del valor m√°s frecuente.
        
        column_name_to_split: int, default=None
            Nombre de la columna a separar si es relevante.

        Retorna
        -------
        cleaned_data: Dataframe
            Dataframe de la data limpia.
        """
        # Crear carpeta EDA/clean_data
        clean_data_folder = os.path.join(self.output_folder, "clean_data")
        os.makedirs(clean_data_folder, exist_ok=True)

        # Copiar el DataFrame para evitar modificar el original
        cleaned_data = self.data.copy()
        
        # Definir una funci√≥n que permita guardar el procesamiento dentro de un FunctionTransformer
        def cleaning_data(data):
            # Separar columnas no at√≥micas
            if column_name_to_split:
                transformer = FunctionTransformer(lambda x: x.str.split(r'[(*?:]', expand=True))
                split_columns = transformer.transform(data[column_name_to_split])
                # Renombrar las columnas seg√∫n la separaci√≥n
                split_columns.columns = column_name_to_split.split('-')

                # Reemplazar 'NaN' con NaN real en las columnas separadas
                split_columns = split_columns.replace('NaN', np.nan)
                split_columns = split_columns.replace('nan', np.nan)
                split_columns = split_columns.replace('0.0', np.nan)

                # Eliminar columna a dividir del Dataframe original
                data.drop(column_name_to_split, axis=1, inplace=True)

                # Incorporar las columnas divididas al Dataframe original
                data = pd.concat([data, split_columns], axis=1)

            # Transformar columnas del dataset a num√©ricas y categ√≥ricas seg√∫n corresponda
            data = self.dtype_transform(data, max_categoric_value)

            # Eliminar valores duplicados
            data.drop_duplicates(inplace=True)

            # Tratar valores nulos
            if null_handling_method == "drop":
                if columns_of_interest is not None:
                    data.dropna(subset=columns_of_interest, inplace=True)
                else:
                    data.dropna(inplace=True)

            elif null_handling_method == "impute":
                data = self.get_imputer(data)
                data.select_dtypes(include='number').fillna(0, inplace=True)
                #data.select_dtypes(include='category').fillna('No Value', inplace=True)
                #data.fillna(0, inplace=True)

            else:
                print("M√©todo de tratamiento de nulos no v√°lido.")
                return
            
            return data

        # Crear un FunctionTransformer para realizar la limpieza del dataset
        data_cleaner = FunctionTransformer(cleaning_data)
        cleaned_data = data_cleaner.fit_transform(cleaned_data)
        cleaned_data = pd.DataFrame(cleaned_data)

        # Guardar los datos procesados en formato .csv en la carpeta EDA/clean_data
        cleaned_data.to_csv(os.path.join(clean_data_folder, "data.csv"), index=False)

        # Crear atributos para poder usarlo en otros m√©todos
        self.clean_data_transformer = data_cleaner # Trasnformador
        self.data_cleaned = cleaned_data # Dataframe con datos limpios
        self.max_categoric_value = max_categoric_value # Valor de camtidad de categor√≠as

        return cleaned_data

# Scale
    def scale(self, categorical_scaler=None, numeric_scaler_stack=False, *numeric_scaler):
        """
        Realiza el escalamiento de un conjunto de datos almacenado en un archivo CSV. Los datos se leen desde el archivo
        y se aplican transformaciones.
        Aplica escaladores logar√≠tmicos y MinMax a las columnas num√©ricas y, opcionalmente, un codificador OneHot a las columnas categ√≥ricas.
        Retorna un transformador que puede ser utilizado posteriormente en otros m√©todos.

        Par√°metros
        ----------
        categorical_scaler: str, {'OneHotEncoder', 'OrdinalEncoder', None}, default=None
            Indica si se usar√° o no un encoder para codificar datos categ√≥ricos.

        numeric_scaler_stack: bool, default=False 
            Indica si deben apilarse los escaladores num√©ricos que se ingresen como input sobre los escaladores existentes en la funci√≥n.

        *numeric_scaler: *arg
            Argumento que permite ingresar escaladores num√©ricos de la libreria Scikit-learn.
            Ejemplo: RobustScaler()

        Retorna
        -------
        - preprocessor: ColumnTransformer
            Transformer que realiza el escalamiento de los datos. Puede ser utilizado en otros m√©todos.
        """

        # Crear la carpeta EDA_fecha/scale
        scale_folder = os.path.join(self.output_folder, "scale")
        os.makedirs(scale_folder, exist_ok=True)

        # Crear una copia de la data original ya limpia
        data = self.data_cleaned.copy()

        # Transformar columnas del dataset a categ√≥ricas seg√∫n corresponda
        #data = self.dtype_transform(data, max_categoric_value)

        # Obtener columnas num√©ricas y categ√≥ricas
        num_cols = data.select_dtypes(include='number').columns
        cat_cols = data.select_dtypes(include='category').columns

        # Obtener columnas a transformar con escaladores num√©ricos
        numeric_features = num_cols.tolist()

        # Escalador Logar√≠tmico y MinMaxScaler
        numeric_transformer = Pipeline(steps=[('scaler', StandardScaler()),
                                            ('minmax', MinMaxScaler())])

        transformers = [('num', numeric_transformer, numeric_features)]

        # Si se introduce alg√∫n *numeric_scaler
        if len(numeric_scaler) > 0:
            if numeric_scaler_stack == True:
                # Agregamos los escaladores al pipeline numeric_transformer
                pasos = [('scaler', StandardScaler()),
                        ('minmax', MinMaxScaler())]
                for scaler in numeric_scaler:
                    step_name = f'{scaler}'
                    pasos.append((step_name, scaler))
                numeric_transformer = Pipeline(steps=pasos)
                transformers = [('num', numeric_transformer, numeric_features)]
            elif numeric_scaler_stack == False:
                # Agregamos los escaladores al un nuevo pipeline
                for scaler in numeric_scaler:
                    step_name = f'{scaler}'
                    new_numeric_transformer = Pipeline(steps=[(step_name, scaler)])
                    transformers.append((step_name, new_numeric_transformer, numeric_features))


        # Si categorical_scaler es True
        if categorical_scaler is not None:
            categorical_features = cat_cols.tolist()
            if categorical_scaler == 'OneHotEncoder':
                # Agregamos OneHotEncoder al pipeline
                categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(drop='if_binary', handle_unknown='ignore'))])
                transformers.append(('cat', categorical_transformer, categorical_features))
            
            elif categorical_scaler == 'OrdinalEncoder':
                # Agregamos OneHotEncoder al pipeline
                categorical_transformer = Pipeline(steps=[('ordinal', OrdinalEncoder())])
                transformers.append(('cat', categorical_transformer, categorical_features))

        # Crear un ColumnTransformer realizar el escalamiento del dataset
        preprocessor = ColumnTransformer(transformers, remainder='drop')
        scaled_data = preprocessor.fit_transform(data)

        # Obtener los nombres de las columnas despu√©s de la transformaci√≥n
        column_names = preprocessor.get_feature_names_out()
        scaled_df = pd.DataFrame(scaled_data, columns=column_names)

        # Crear atributo scale_transformer para poder usarlo en otros m√©todos
        self.scale_transformer = preprocessor

        # Crear atributo para guardar nombre de columnas del dataset escalado
        self.scale_columns = column_names

        # Crear atributo para guardar data escalada
        self.scaled_data = scaled_df

        # Guardar el ColumnTransformer para su uso posterior
        transformer_file = os.path.join(scale_folder, "column_transformer.pkl")
        joblib.dump(preprocessor, transformer_file)

        # Guardar los datos escalados en un archivo CSV
        scaled_df.to_csv(os.path.join(scale_folder, "scaled_features.csv"), index=False)

        return preprocessor

# Make clusters
    def make_clusters(self, clustering_algorithm="KMeans", max_clusters=10):
        """
        Funciona a partir de la data previamente escalada con el m√©todo scale, por lo que los m√©todos clean_data y scale
        deben haber sido utilizados previamente.

        Mediante un pipeline, se llaman a los transformers generados en los m√©todos clean_data y scale, adem√°s de aplicar el algor√≠tmos
        de custerizaci√≥n indicado en el input del m√©todo.

        Clusteriza los datos entregados y crea una columna con las etiquetas de clusters en el dataset.

        Genera un gr√°fico con el m√©todo del codo, a partir de la cual se debe indicar en un input la cantidad de clusters deseadas para realizar la clusterizaci√≥n de los datos.

        Crea un gr√°fico en 2 dimensiones de los datos clusterizados y diferenciados cada cluster por color.

        Par√°metros
        ----------
        clustering_algorithm: {'AffinityPropagation', 'AgglomerativeClustering', 'Birch', 'KMeans', 'MiniBatchKMeans', 'MeanShift', 'SpectralClustering'}, default='KMeans'. 
            Algor√≠tmo de clustering deseado.

        max_clusters: int, default=10. 
            Cantidad m√°xima de clusters admitida.
        """
        
        # Hacer una copia del dataset original
        data = self.data.copy()

        # Crear el pipeline con las etapas de limpieza y escalado
        preprocessing_pipeline = Pipeline([
            ('clean_data', self.clean_data_transformer),
            ('scale', self.scale_transformer)
        ])

        # Aplicar el pipeline a los datos crudos
        cleaned_and_scaled_data = preprocessing_pipeline.fit_transform(data)

        # Convertir el resultado a un DataFrame
        cleaned_and_scaled_data = pd.DataFrame(cleaned_and_scaled_data, columns=self.scale_columns)

        # Crear la carpeta EDA_fecha/clusters
        clusters_folder = os.path.join(self.output_folder, "clusters")
        os.makedirs(clusters_folder, exist_ok=True)

        # Realizar m√©todo del codo para definir cantidad de clusters a aplicar
        # Inicializar una lista para almacenar las sumas de cuadrados intra-cluster
        wcss = []
            
        # Evaluar diferentes n√∫meros de clusters
        for i in range(1, max_clusters + 1):
            kmeans = KMeans(n_clusters=i, random_state=42)
            kmeans.fit(cleaned_and_scaled_data)
            # Calcular la suma de cuadrados intra-cluster y agregarla a la lista
            wcss.append(kmeans.inertia_)

        # Graficar el m√©todo del codo para seleccionar el n√∫mero √≥ptimo de clusters
        plt.figure(figsize=(8, 6))
        plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
        plt.xlabel('N√∫mero de Clusters')
        plt.ylabel('Suma de Cuadrados Intra-cluster (WCSS)')
        plt.title('M√©todo del Codo')
        plt.savefig(os.path.join(clusters_folder, "elbow_method.png"))

        plt.show()

        # Pedir al usuario que ingrese el n√∫mero deseado de clusters
        selected_num_clusters = int(input("Ingresa el n√∫mero de clusters deseado: "))

        # Definir el modelo a entrenar seg√∫n input del m√©todo
        if clustering_algorithm == 'MiniBatchKMeans':
            algorithm = MiniBatchKMeans(n_clusters=selected_num_clusters, random_state=42)

        elif clustering_algorithm == 'SpectralClustering':
            algorithm = SpectralClustering(n_clusters=selected_num_clusters, assign_labels='kmeans', random_state=42)

        elif clustering_algorithm == 'Birch':
            algorithm = Birch(n_clusters=selected_num_clusters)

        elif clustering_algorithm == 'AgglomerativeClustering':
            algorithm = AgglomerativeClustering(n_clusters=selected_num_clusters)

        elif clustering_algorithm == 'AffinityPropagation':
            algorithm = AffinityPropagation(random_state=42)
        
        elif clustering_algorithm == 'MeanShift':
            algorithm = MeanShift()
        
        else:
            algorithm = KMeans(n_clusters=selected_num_clusters, random_state=42)

        # Crear pipeline con preprocesamiento y clusterizaci√≥n
        clustering_pipe = Pipeline([
            ('preprocessing',preprocessing_pipeline),
            ('clustering', algorithm)
        ])

        # Hacer una nueva copia del dataset original y 
        data = self.data.copy()
        # Aplicar pipeline
        cluster_labels = clustering_pipe.fit_predict(data)

        # Guardar los datos con su respectivo cluster
        cluster_data = cleaned_and_scaled_data.copy()
        cluster_data["Cluster"] = cluster_labels
        cluster_data.to_csv(os.path.join(clusters_folder, "data_clusters.csv"), index=False)

        # Reducci√≥n de dimensionalidad y proyecci√≥n de datos en 2D
        pca = PCA(n_components=2)
        reduced_data = pca.fit_transform(cleaned_and_scaled_data)

        # Graficar la proyecci√≥n de datos coloreada por cluster
        plt.figure(figsize=(8, 6))
        plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=cluster_labels)
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.title('Proyecci√≥n de Datos con Coloreo por Cluster')
        plt.savefig(os.path.join(clusters_folder, "data_projection.png"))

        # Mostrar y guardar los gr√°ficos
        plt.show()

# Detect anomalies
    def detect_anomalies(self, algorithm):
        """
        Funciona a partir de la data previamente escalada con el m√©todo scale, por lo que los m√©todos clean_data y scale
        deben haber sido utilizados previamente.

        Mediante un pipeline, se llaman a los transformers generados en los m√©todos clean_data y scale, adem√°s de aplicar el algoritmo
        de clustering indicado en el input del m√©todo.

        Detecta anomal√≠as en los datos entregados y crea una columna con las etiquetas de anomal√≠as en el dataset.

        Genera un gr√°fico con el m√©todo del codo, a partir del cual se debe indicar en un input la
        cantidad de clusters deseadas para realizar la clusterizaci√≥n de los datos.

        Par√°metros
        ----------
        algorithm: estimator, default=IsolationForest()
            Algoritmo de detecci√≥n de anomal√≠as de scikit-learn, por ejemplo, OneClassSVM, IsolationForest, EllipticEnvelope, etc.
        """
        # Crear la carpeta EDA_fecha/anomalies
        anomalies_folder = os.path.join(self.output_folder, "anomalies")
        os.makedirs(anomalies_folder, exist_ok=True)
        
        # Hacer una copia del dataset original
        data = self.data.copy()

        # Crear el pipeline con las etapas de limpieza y escalado
        preprocessing_pipeline = Pipeline([
            ('clean_data', self.clean_data_transformer),
            ('scale', self.scale_transformer)
        ])

        # Aplicar el pipeline a los datos crudos
        cleaned_and_scaled_data = preprocessing_pipeline.fit_transform(data)

        # Convertir el resultado a un DataFrame
        cleaned_and_scaled_data = pd.DataFrame(cleaned_and_scaled_data, columns=self.scale_columns)

        if algorithm is None:
            algorithm = IsolationForest()

        # Entrenar el detector de anomal√≠as
        anomaly_detector = algorithm.fit(cleaned_and_scaled_data)

        # Predecir las anomal√≠as en los datos
        anomaly_labels = anomaly_detector.predict(cleaned_and_scaled_data)

        # Guardar los datos con etiquetas de anomal√≠as en formato CSV
        anomalies_data = cleaned_and_scaled_data.copy()
        anomalies_data['anomaly_label'] = anomaly_labels
        anomalies_file = os.path.join(anomalies_folder, "data_anomalies.csv")
        anomalies_data.to_csv(anomalies_file, index=False)

        # Proyectar los datos en 2D para graficar
        pca = PCA(n_components=2)  # Proyecci√≥n a 2 dimensiones para graficar
        projected_data = pca.fit_transform(cleaned_and_scaled_data)

        # Graficar los datos coloreando por las etiquetas de anomal√≠as
        plt.figure(figsize=(8, 6))
        plt.scatter(projected_data[:, 0], projected_data[:, 1], c=anomaly_labels, cmap='viridis')
        plt.title('Detecci√≥n de Anomal√≠as')
        plt.xlabel('Dimensi√≥n 1')
        plt.ylabel('Dimensi√≥n 2')
        plt.colorbar()
        plt.savefig(os.path.join(anomalies_folder, 'anomalies_plot.png'))
        plt.close()


# Profile
    def profile(self, clusters_algorithm='KMeans',anomalies_algorithm=None):
        '''
        Llama a todos los m√©todos de la clase profiler en orden.
        '''
        # Llamar a los m√©todos en el orden deseado
        self.summarize()
        self.plot_vars()
        self.clean_data()
        self.scale()
        self.make_clusters(clusters_algorithm)
        if anomalies_algorithm:
            self.detect_anomalies(anomalies_algorithm)


# clearGarbage
    def clearGarbage(self):
        """
        Elimina todos los archivos y carpetas existentes dentro del directorio de trabajo.
        """
        if os.path.exists(self.output_folder):
            # Elimina el directorio de salida y su contenido de forma recursiva
            shutil.rmtree(self.output_folder)
            print(f"Se ha eliminado el contenido de {self.output_folder}")
        else:
            print(f"El directorio {self.output_folder} no existe.")
````
