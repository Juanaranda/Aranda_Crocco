El objetivo de este estudio es predecir la probabilidad de incumplimiento de pago de los clientes de Giturra. Para esto, se busca identificar aquellos clientes que tienen un mayor riesgo de no pagar sus préstamos. Esto lo ayudará a tomar decisiones más informadas y ofrecer condiciones más favorables a los clientes con menor riesgo de impago, reduciendo así los préstamos otorgados a individuos con mayor probabilidad de incumplir.

Estos datos se utilizarán para entrenar los modelos y predecir la probabilidad de incumplimiento de pago de los clientes.


Dado que los datos presentan desbalanceo, es decir, una proporción desigual entre las clases de interés (pagadores e incumplidores), el uso de la métrica 'accuracy' no sería adecuado, ya que puede dar una impresión engañosa de precisión debido al desequilibrio en las clases.

En este contexto, se podría enfocar en métricas como precision, recall o f1-score, que están más centradas en el rendimiento del modelo para predecir la clase minoritaria, es decir, los clientes que incumplen con los pagos. La elección específica entre estas métricas dependerá de la estrategia del banco y sus prioridades.

Si Giturra busca minimizar los falsos positivos (predecir que un cliente pagará pero en realidad no lo hace), la precision sería crucial. Si el énfasis está en identificar la mayor cantidad posible de clientes que podrían incumplir (aunque se incluyan algunos falsos positivos), recall sería más relevante. El f1-score es una medida que combina precisión y recall y podría ser una opción equilibrada si se busca un compromiso entre ambas.

La métrica escogida debe alinearse con los objetivos del banco, considerando la importancia relativa de predecir correctamente los clientes que incumplirán con los pagos frente a evitar falsas alarmas sobre aquellos que pagarán, lo que significa un costo de oportunidad.


Finalmente, para abordar este desafío se implementaron dos modelos: un clasificador SVM (Support Vector Machine) a través del algoritmo SVC (Support Vector Classifier) y un clasificador LightGBM (Light Gradient Boosting Machine). Los datos se prepararon realizando una limpieza que incluyó la eliminación de valores nulos e inconsistencias, junto con la creación de nuevas variables como la relación entre ingresos y deuda, para capturar relaciones significativas. Además, se aplicó escalado y codificación one-hot para representar adecuadamente las variables en los modelos. Para manejar el desbalance de clases, se implementó Random Over Sampling, generando réplicas de la clase minoritaria. Estos pasos en el preprocesamiento mejoraron el desempeño de los modelos, especialmente en la detección de clientes fraudulentos.

El mejor modelo implementado, LGBMClassifier, alcanzó un F1 score de 0.5982, demostrando mejoras significativas sobre el baseline. ## 3. Preparación de Datos


###Limpieza de datos

Antes de llevar a cabo el preprocesamiento de los datos, se debió realizar una limpieza minuciosa de ellos. En primer lugar, se eliminaron las columnas que no aportan información relevante al problema como el identificador del cliente y el salario mensual de este. En particular, esta última columna se descartó debido a que entrega la misma información que la columna que representa los ingresos anuales de los clientes.<br>
<br>
Luego, se trataron las columnas con datos inconsistentes, por ejemplo, las edades negativas y mayores a 100, tasas de interés por sobre el 100%, entre otros.<br>
<br>
Por último, se abordó el problema de las columnas con datos faltantes. Debido a que los datos faltantes no representan un procentaje significativo del todal de los datos, se optó por su eliminación.

```python
def clean_data(df):
    df = df.copy()

    columns_to_drop = ['monthly_inhand_salary', # duplicidad de información con el salario anual
                       'total_emi_per_month', # baja correlacion
                       'customer_id' # información irrelevante
                       ]
    df = df.drop(columns_to_drop,axis=1)

    df = df[(df.age >= 16) & (df.age<=100)] # Eliminar edades muy altas o muy bajas para poder abrir una cuenta
    df = df[(df.num_bank_accounts >= 1) & (df.num_bank_accounts <= 30)] # Eliminar valores que no concuerdan con ser clientes del banco, es decir, cantidad de cuentas negativas o cero o muy altas
    df = df[(df.num_credit_card <= 30)]
    df = df[(df.interest_rate <= 100)] # Eliminar tasas mayores a un 100%
    df = df[(df.num_of_loan >= 0) & (df.num_of_loan<=50)]
    df = df[(df.num_of_delayed_payment <= 100)]
    df = df[(df.num_credit_inquiries <= 100)]


    df = df.dropna() # Eliminar valores nulos

    return df

df_clean = clean_data(df)
eda(df_clean)
```

A continuación, se procedió a modificar los tipos de las columnas mal leídas. Además, se decidió pasar los datos con menos de 10 valores únicos y de tipo object a tipo categórico.

```python
cat_cols = ['payment_of_min_amount','payment_behaviour','occupation','credit_score']
df_clean[cat_cols] = df_clean[cat_cols].astype("category")
```

####Featuring engineer

Se decidió crear las siguientes variables adicionales para brindar una mejor orientación a los modelos. Las variables creadas son las siguientes:

- **Relación entre la deuda y el ingreso anual (debt_rate):** Se construyó esta variable conocida como tasa de endeudameinto, o debt rate en inglés. Esta permite entender la carga financiera de una persona en relación con sus ingresos.

```python
df_clean['debt_ratio'] = df_clean['outstanding_debt']/df_clean['annual_income']
```

- **Relación entre la edad y el ingreso anual (age_income_ratio):** Se construyó esta variable con el fin de conocer cómo evoluciona el ingreso de los clientes en función de sus edades. Esto permite conocer posibles relaciones entre los ingresos y la experiencia laboral, o la estabilidad financiera del cliente en diferentes etapas de la vida.

```python
df_clean['age_income_ratio'] = df_clean['annual_income']/df_clean['age']
```

- **Agrupación de edades por rango (age_category)**: Se decidió crear una variable que desagregue a la edad, con el fin de tratarla como una variable categórica. Esto puede facilitar la comprensión e interpretación de las tendencias generales de los datos respecto a las edades de los clientes.

```python
bins = [16, 24, 34, 44, 60]
labels = ['jovenes', 'adultos_jovenes', 'adultez_mediana', 'mayores']

# Crear una nueva característica de categorías de edad
df_clean['age_category'] = pd.cut(df_clean['age'], bins=bins, labels=labels, include_lowest=True, right=False)
df_clean['age_category'].cat.add_categories('otro', inplace=True)
df_clean['age_category'].fillna('otro', inplace=True)
df_clean['age_category'] = df_clean['age_category'].astype("category")
```

## 4. Baseline

Se entrenaron ocho modelos de clasificación distintos para predecir la probabilidad de que los clientes no fueran confiables (o fraudulentos). Estos modelos incluyeron varios algoritmos y métodos para comparar y evaluar su desempeño. Entre los modelos se incluyó uno llamado "dummy", el cual se realiza en base de una estrategia estratificada, la cual se utilizó como punto de referencia inicial. Este modelo simplemente asigna las clases de forma aleatoria según la distribución de las clases en los datos de entrenamiento, sin hacer inferencias basadas en los datos.<br>
<br>
Además, se utilizó la métrica F1, que combina la precisión y el recall en una sola medida, para evaluar el desempeño de los modelos. Dado que toma en cuenta tanto los falsos positivos como los falsos negativos, esta métrica es ideal para el problema de detección de clientes fraudulentos. Para corregir la disparidad entre las clases de clientes fraudulentos y no fraudulentos, se utilizó la técnica de Random Over Sampling. Esta estrategia produjo replicas de la clase minoritaria, lo que permitió que las clases se distribuyeran de manera equitativa en el conjunto de datos de entrenamiento, lo que mejoró la capacidad de los modelos para detectar correctamente casos de fraude.

```python
# Crear las instancias de los clasificadores
classifiers = {
    'Dummy': DummyClassifier(strategy='stratified', random_state=42),
    'LogisticRegression': LogisticRegression(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'SVC': SVC(),
    'RandomForestClassifier': RandomForestClassifier(),
    'LGBMClassifier': LGBMClassifier(),
    'XGBClassifier': XGBClassifier()
}

# Crear un diccionario para almacenar las métricas de cada modelo
metrics = {'Model': [], 'F1-Score': []}

# Iterar sobre cada clasificador, entrenar, evaluar y guardar la métrica F1-score
for clf_name, clf in classifiers.items():
    #pipeline = Pipeline([
        #('preprocessor', preprocessor),  # Utiliza el ColumnTransformer definido anteriormente
        #('classifier', clf)  # El clasificador actual
    #])
    pipeline = ImbPipeline([
        ('preprocessor', preprocessor),
        ('sampler', RandomOverSampler()),  # Balanceo de clases
        ('classifier', clf)
    ])

    # Entrenar el modelo
    pipeline.fit(X_train, y_train)

    # Realizar predicciones
    predictions = pipeline.predict(X_test)

    # Calcular F1-Score y guardar en el diccionario de métricas
    #f1 = f1_score(y_test, predictions, average='weighted')  # Cambia el average si es necesario
    f1 = f1_score(y_test, predictions, pos_label=1) #Para ver la clase positiva
    metrics['Model'].append(clf_name)
    metrics['F1-Score'].append(f1)

    # Imprimir reporte de clasificación
    print(f"Classification report for {clf_name}:")
    print(classification_report(y_test, predictions))
    print('-' * 50)

# Convertir el diccionario de métricas a un DataFrame y ordenar por F1-Score de mayor a menor
import pandas as pd
metrics_df = pd.DataFrame(metrics).sort_values(by='F1-Score', ascending=False)
print("Modelos ordenados por F1-Score:")
print(metrics_df)
```

## 5. Optimización del Modelo

Debido a que los resultados de la parte anterior arrojaron un mejor rendimiento de los modelos SVC y LightGBM, se optó por utilizar estos dos en el modelado final. En busca de mejorar el rendimiento de ambos modelos, se realizó una optimización de hiperparámetros, empleando las tácnicas de HalvingGridSearch y Optuna.

A continuación se puede ver el proceso para la optimización de hiperparámetros utilizando HalvingGridSearch:

```python
# Pipeline con LGBMClassifier para optimizacion
modelo_lgbm = LGBMClassifier(random_state=seed, verbose=-1)
pipeline_lgbm = ImbPipeline([
                            ('preprocessor', preprocessor),
                            ('sampler', RandomOverSampler()),
                            ('classifier', modelo_lgbm)
                            ])

# Pipeline con SVC para optimizacion
modelo_svc = SVC(random_state=seed)
pipeline_svc = ImbPipeline([
                        ('preprocessor', preprocessor),
                        ('sampler', RandomOverSampler()),
                        ('classifier', modelo_svc)
                        ])

# Definición de la grillaa de parámetros
param_grid_lgbm = {
    'classifier__num_leaves': [10, 20, 30],
    'classifier__learning_rate': [0.1, 0.05, 0.01],
    'classifier__reg_lambda': [0.1, 0.5, 1.0],
    'classifier__reg_alpha': [0.1, 0.5, 1.0]
}

param_grid_svc = {
    'classifier__C': [1, 10, 100],
    'classifier__kernel': ['linear', 'rbf']
}

# Se crea instancia de HalvingGridSearch para el modelo LightGBM
halving_search_lgbm = HalvingGridSearchCV(
    pipeline_lgbm, param_grid_lgbm, scoring='f1', verbose=1,n_jobs=-1
)
halving_search_lgbm.fit(X_train, y_train)

# Se imprimen los mejores resultados para LightGBM
print("Mejores hiperparámetros para LGBMClassifier:")
print(halving_search_lgbm.best_params_)
print("Mejor puntuación F1:", halving_search_lgbm.best_score_)

# Se crea instancia de HalvingGridSearch para el modelo SVM
halving_search_svc = HalvingGridSearchCV(
    pipeline_svc, param_grid_svc, scoring='f1', verbose=1,n_jobs=-1
)
halving_search_svc.fit(X_train, y_train)

# Se imprimen los mejores resultados para SVC
print("Mejores hiperparámetros para SVC:")
print(halving_search_svc.best_params_)
print("Mejor puntuación F1:", halving_search_svc.best_score_)
```

```python
def evaluar_modelo(modelo, X_test, y_test):

    print(modelo.named_steps['classifier'].__class__.__name__)
    y_pred = modelo.predict(X_test)
    reporte = classification_report(y_test, y_pred)
    print(reporte)
    y_pred = modelo.predict(X_test)
    score = f1_score(y_test, y_pred, pos_label=1) # f1 para encontrar credit score 1, es decir, los riesgoso
    return (modelo.named_steps['classifier'].__class__.__name__,score)
```

```python
metricas_gridsearch = []

best_pipeline_lgbm = halving_search_lgbm.best_estimator_
best_pipeline_svc = halving_search_svc.best_estimator_

metricas_gridsearch.append(evaluar_modelo(best_pipeline_lgbm, X_test, y_test))
metricas_gridsearch.append(evaluar_modelo(best_pipeline_svc, X_test, y_test))

for metrica in metricas_gridsearch:
    print(f"Modelo {metrica[0]} - F1 Score {round(metrica[1], 4)}.")
```

 Esta metodología iterativa permitió explorar de manera exhaustiva la grilla de hiperparámetros, seleccionando progresivamente las combinaciones más prometedoras en cada iteración. Esta estrategia redujo significativamente la cantidad de configuraciones evaluadas en comparación con un enfoque de búsqueda en toda la grilla. Se puede apreciar que el modelo que presentó un mejor rendimiento en cuanto a la métrica de *F1 Score* fue LightGBM.<br>

A continuación se puede ver el proceso para la optimización de hiperparámetros utilizando Optuna:

```python
def objective(trial,pipeline,X_train, y_train,X_test,y_test):
    model_name = pipeline.named_steps['classifier'].__class__.__name__

    if model_name == "SVC":
        params = {
            'C': trial.suggest_float("C", 0.01, 20.0),
            'kernel': trial.suggest_categorical("kernel", ["linear", "rbf"]),
            'gamma': trial.suggest_categorical("gamma", ["scale", "auto"])
        }
    elif model_name == "LGBMClassifier":
        params = {
            'boosting_type': trial.suggest_categorical("boosting_type", ["gbdt", "dart"]),
            'n_estimators': trial.suggest_int("n_estimators", 500, 750),
            'max_depth': trial.suggest_int("max_depth", 10, 50),
            'learning_rate': trial.suggest_float("learning_rate", 0.01, 0.2),
            'reg_alpha': trial.suggest_float('reg_alpha', 0, 10.0),
            'reg_lambda': trial.suggest_float('reg_lambda', 0, 10.0),
            'num_leaves': trial.suggest_int('num_leaves', 1, 100)

        }
    else:
        raise ValueError(f"Modelo {model_name} no soportado. Agregar a funcion objetivo.")

    # Actualizar el modelo con los hiperparámetros sugeridos
    pipeline.named_steps['classifier'].set_params(**params)

    # Concatenar los conjuntos de datos
    X = pd.concat([X_train, X_test], axis=0, ignore_index=True)
    y = pd.concat([y_train, y_test], axis=0, ignore_index=True)

     # Calcular la puntuación de validación cruzada
    scores = cross_val_score(pipeline, X, y, scoring='f1', cv=5)

    # Calcular la media de las puntuaciones de validación cruzada
    score = scores.mean()


    return score


def print_best_callback(study, trial):
    # print(f"Trial {trial.number}, Best value: {study.best_value}", end='\r')
    print(f"Trial {trial.number}, Best value: {study.best_value}")


def optimize_model(pipeline,n_trials,X_train, y_train,X_test,y_test):
    best_loss =  float('inf')
    print(f'''Optimizando modelo {pipeline.named_steps['classifier'].__class__.__name__}.''')
    tiempo_inicio = time.time()
    study = optuna.create_study(direction='maximize',sampler=optuna.samplers.TPESampler(seed=seed))
    # study.optimize(objective, n_trials=n_trials, callbacks=[print_best_callback])

    study.optimize(lambda trial: objective(trial, pipeline, X_train, y_train, X_test, y_test), n_trials=n_trials, callbacks=[print_best_callback])
    tiempo_fin = time.time()
    tiempo_transcurrido = round(tiempo_fin - tiempo_inicio, 2)
    print(f"Optimizacion finalizada para {pipeline.named_steps['classifier'].__class__.__name__}, Best value: {study.best_value}. Mejor resultado obtenido en la trial #{study.best_trial.number}.\n Tiempo total {tiempo_transcurrido} segundos.\n")

    best_params = study.best_params

    return best_params
```

```python
metricas_optuna = []

modelo_lgbm = LGBMClassifier(random_state=seed, verbose=-1)
pipeline_lgbm = ImbPipeline([
                        ('preprocessor', preprocessor),
                        ('sampler', RandomOverSampler()),
                        ('classifier', modelo_lgbm)
                        ])

modelo_svc = SVC(random_state=seed)
pipeline_svc = ImbPipeline([
                        ('preprocessor', preprocessor),
                        ('sampler', RandomOverSampler()),
                        ('classifier', modelo_svc)
                        ])
trials = 10

best_params_lgbm = optimize_model(pipeline_lgbm,trials,X_train, y_train,X_test,y_test)
pipeline_lgbm.named_steps['classifier'].set_params(**best_params_lgbm)
pipeline_lgbm.fit(X_train, y_train)
metricas_optuna.append(evaluar_modelo(pipeline_lgbm,X_test,y_test))

best_params_svc = optimize_model(pipeline_svc,trials,X_train, y_train,X_test,y_test)
pipeline_svc.named_steps['classifier'].set_params(**best_params_svc)
pipeline_svc.fit(X_train, y_train)
metricas_optuna.append(evaluar_modelo(pipeline_svc,X_test,y_test))


for metrica in metricas_optuna:
    print(f"Modelo {metrica[0]} - F1 Score {round(metrica[1], 4)}.")
```

Optuna, basado en el método de exploración bayesiana, realiza una búsqueda inteligente en el espacio de hiperparámetros. Esta herramienta busca de manera estratégica las mejores configuraciones al analizar cuidadosamente las evaluaciones de los modelos en cada iteración. Al hacerlo, permite identificar y seleccionar las combinaciones más óptimas de hiperparámetros para mejorar el rendimiento de los modelos. Nuevamente, luego de aplicar este método, el modelo LightGBM es el que muestra los mejores resultados en cuanto a la métrica del *F1-score*.

Observando los resultados, se hace evidente que el modelo LightGBM, utilizando la optimización de hiperparámetros con HalvingGridSearch, demostró ser el mejor de los modelos analizados. Este modelo final se elige como el mejor para hacer predicciones de clientes falsas. La capacidad de LightGBM para capturar patrones y relaciones no lineales complejos, así como su tiempo de entrenamiento y optimización de hiperparámetros eficientes, son dos factores que justifican su elección.
